# Challenge 2

```python
titles_author = '''
SELECT
authors.au_id AS 'AUTHOR ID',
authors.au_lname AS 'LAST NAME',
authors.au_fname AS 'FIRST NAME', 
pub.pub_name AS 'PUBLISHER', COUNT(*) AS 'TITLE COUNT'
FROM authors
INNER JOIN titleauthor ta ON authors.au_id = ta.au_id 
INNER JOIN titles title ON ta.title_id = title.title_id
INNER JOIN publishers pub ON title.pub_id = pub.pub_id
GROUP BY title
'''
```

```python
titles_author = pd.read_sql_query(titles_author, engine)
```

```python
titles_author
```

```python
verification = titles_author["TITLE COUNT"].sum()
verification
```

```python
top_3 = '''
SELECT
a.au_id AS 'AUTHOR ID',
a.au_lname AS 'LAST NAME',
a.au_fname AS 'FIRST NAME',
SUM(ytd_sales) AS 'TOTAL'
FROM titleauthor ta
INNER JOIN authors a ON a.au_id = ta.au_id 
INNER JOIN titles ti ON ti.title_id = ta.title_id
GROUP BY a.au_id, a.au_lname, a.au_fname
ORDER BY TOTAL DESC
LIMIT 3
'''
#aquí no he tenido en cuenta los autores que no han escrito ni vendido ningún libro porque solo queríamos sacar los top3
```

```python
top_3 = pd.read_sql_query(top_3, engine)
top_3
```

```python
a = '''
select * 
from authors a
left join titleauthor ti on a.au_id=ti.au_id
'''
authors = pd.read_sql_query(a, engine)
authors
# he juntado las dos tablas para ver la columna en común, porque en el ejercicio del challenge 3 
# no he tenido en cuenta los autores que no han escrito ningún libro y por tanto no han vendido ninguno
```

```python
ranking_authors= '''
SELECT
a.au_id AS 'AUTHOR ID',
a.au_lname AS 'LAST NAME',
a.au_fname AS 'FIRST NAME',
IFNULL(t.TOTAL, 0) AS TOTAL
FROM authors a 
LEFT JOIN (
    SELECT
    a.au_id,
    a.au_lname AS 'LAST NAME',
    a.au_fname AS 'FIRST NAME',
    SUM(ytd_sales) AS 'TOTAL'
    FROM titleauthor ta
    INNER JOIN authors a ON a.au_id = ta.au_id 
    INNER JOIN titles ti ON ti.title_id = ta.title_id
    GROUP BY a.au_id, a.au_lname, a.au_fname
    ) t ON t.au_id = a.au_id
    ORDER BY TOTAL DESC
'''
#he tenido que hacer una subquery por como tenía la primera query y así sacar también los escritores que han vendido 0 titles
```

```python
ranking_authors = pd.read_sql_query(ranking_authors, engine)
ranking_authors
```
