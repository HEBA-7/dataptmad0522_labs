```python
temporal_query_2 = """ 
CREATE TEMPORARY TABLE AS TOP_AU_PROFIT
WITH r (AUTHOR_ID, TITLE_ID, ADVANCE, ROYALTY) AS
    (SELECT 
    s.title_id AS 'TITLE_ID',
    ta.au_id AS 'AUTHOR_ID',
    ti.advance * ta.royaltyper / 100 AS 'ADVANCE',
    ti.price * s.qty * ti.royalty / 100 * ta.royaltyper / 100 AS 'ROYALTY'

    FROM Sales s
    LEFT JOIN titles ti ON ti.title_id = s.title_id
    LEFT JOIN titleauthor ta ON ta.title_id = s.title_id),

y (TITLE_ID, AUTHOR_ID, AGG_ROYALTY, AGG_ADVANCE)
AS 
    (SELECT 
    r.TITLE_ID, 
    r.AUTHOR_ID,
    SUM(r.ROYALTY) AS 'AGG_ROYALTY',
    SUM (r.ADVANCE) AS 'AGG_ADVANCE'
    FROM r
    GROUP BY r.AUTHOR_ID, r.TITLE_ID),

x (AUTHOR_ID, PROFIT)
AS
    (SELECT
    y.AUTHOR_ID,
    y.AGG_ROYALTY + y.AGG_ADVANCE AS PROFIT
    FROM y)

SELECT
x.AUTHOR_ID AS 'AUTHOR_ID',
SUM(x.PROFIT) AS 'TOTAL_PROFIT'
FROM x
GROUP BY AUTHOR_ID
ORDER BY TOTAL_PROFIT DESC
LIMIT 3
    
"""
```

```python
final_query = """ 
WITH r (AUTHOR_ID, TITLE_ID, ADVANCE, ROYALTY) AS
    (SELECT 
    s.title_id AS 'TITLE_ID',
    ta.au_id AS 'AUTHOR_ID',
    ti.advance * ta.royaltyper / 100 AS 'ADVANCE',
    ti.price * s.qty * ti.royalty / 100 * ta.royaltyper / 100 AS 'ROYALTY'

    FROM Sales s
    LEFT JOIN titles ti ON ti.title_id = s.title_id
    LEFT JOIN titleauthor ta ON ta.title_id = s.title_id),

y (TITLE_ID, AUTHOR_ID, AGG_ROYALTY, AGG_ADVANCE)
AS 
    (SELECT 
    r.TITLE_ID, 
    r.AUTHOR_ID,
    SUM(r.ROYALTY) AS 'AGG_ROYALTY',
    SUM (r.ADVANCE) AS 'AGG_ADVANCE'
    FROM r
    GROUP BY r.AUTHOR_ID, r.TITLE_ID),

x (AUTHOR_ID, PROFIT)
AS
    (SELECT
    y.AUTHOR_ID,
    y.AGG_ROYALTY + y.AGG_ADVANCE AS PROFIT
    FROM y)

SELECT
x.AUTHOR_ID AS 'AUTHOR_ID',
SUM(x.PROFIT) AS 'TOTAL_PROFIT'
FROM x
GROUP BY AUTHOR_ID
ORDER BY TOTAL_PROFIT DESC
LIMIT 3
"""
    
```
