####  Display the 20 latest earthquakes info (date, time, latitude, longitude and region name) by the EMSC as a pandas dataframe.

````python
# This is the url you will scrape in this exercise
url = 'https://www.emsc-csem.org/Earthquake/'
````

````python
earthquakes = requests.get(url)
print(type(earthquakes))
earthquakes = earthquakes.content
print(type(earthquakes))
earthquakes
````

````python
# your code here
parsed_earth = bs4.BeautifulSoup(earthquakes, "html.parser") 
print(type(parsed_earth))
len(parsed_earth)
````

````python
earth_info = parsed_earth.find("tbody", {"id":"tbody"}).find_all("tr")[:20]
len(earth_info)
print(type(earth_info))
earth_info
````

````python
earth_info_list = []
earth_info_d = {}
for t in earth_info:
    earth_info_d = {}
    earth_info_d["date"] = t.find("td", {"class": "tabev6"}).find_all("a")[0].string.strip()
    earth_info_d["time"] = t.find("td", {"class": "tabev6"}).find_all("i", {"class": "ago"})[0].string.strip()
    earth_info_d["longitude"] = t.find_all("td", {"class": "tabev1"})[1].string.strip()
    earth_info_d["latitude"] = t.find_all("td", {"class": "tabev1"})[0].string.strip()
    earth_info_d["region"] = t.find("td", {"class": "tb_region"}).string.strip()
    earth_info_list.append(earth_info_d)
````

````python
import pandas as pd
df_earth_info = pd.DataFrame(earth_info_list)
df_earth_info
````

#### List all language names and number of related articles in the order they appear in wikipedia.org.

````python
# This is the url you will scrape in this exercise
url = 'https://www.wikipedia.org/'
lang = requests.get(url)
print(type(lang))
lang = lang.content
print(type(lang))
````

````python
# your code here
parsed_lang = bs4.BeautifulSoup(lang, "html.parser") 
print(type(parsed_lang))
len(parsed_lang)
````

````python
num_lang = parsed_lang.find_all("div", {"class": "central-featured-lang"})
num_lang
````

````python
wiki_lang_list = []
for lang in num_lang:
    wiki_speakers_d = {}
    wiki_speakers_d["language"] = lang.find("strong").string.strip()
    wiki_speakers_d["num_lang"] = lang.find("bdi").string.strip()
    wiki_lang_list.append(wiki_speakers_d)
````

````python
df_wiki_lang = pd.DataFrame(wiki_lang_list)
df_wiki_lang
````

#### Display the top 10 languages by number of native speakers stored in a pandas dataframe.

````python
# This is the url you will scrape in this exercise
url = 'https://en.wikipedia.org/wiki/List_of_languages_by_number_of_native_speakers'
wiki = requests.get(url)
print(type(wiki))
wiki = wiki.content
print(type(wiki))
````

````python
# your code here
parsed_wiki = bs4.BeautifulSoup(wiki, "html.parser") 
print(type(parsed_wiki))
len(parsed_wiki)
````

````python
#parsed_wiki
````

````python
wiki_speakers = parsed_wiki.find_all("tbody")[0].find_all("tr")[1:11]
len(wiki_speakers)
wiki_speakers
````

````python
wiki_speakers_list = []
for lang in wiki_speakers:
    wiki_speakers_d = {}
    wiki_speakers_d["language"] = lang.find_all("td")[1].find("a").string.strip()
    wiki_speakers_d["Native Speakers"] = lang.find_all("td")[2].string.strip()
    wiki_speakers_list.append(wiki_speakers_d)
````

````python
import pandas as pd
df_wiki_speakers = pd.DataFrame(wiki_speakers_list)
df_wiki_speakers 
````
